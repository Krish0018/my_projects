## Movie Review Analysis

```python
#let's load a data and do complete analysis and make a model
from nltk.corpus import movie_reviews
nltk.download('movie_reviews')
```

```python
print(movie_reviews.categories())
```

```python
pos_rev = movie_reviews.fileids('pos')
len(pos_rev)
```

```python
neg_rev = movie_reviews.fileids('neg')
len(neg_rev)
#neg_rev
```

```python
rev = nltk.corpus.movie_reviews.words('pos/cv000_29590.txt')
rev
```

```python
#let's clean the review
review_lst1 = []
for review in neg_rev:
    review_text_neg = nltk.corpus.movie_reviews.words(review)
    rev_strng = " ".join(review_text_neg)
    rev_strng = rev_strng.replace(' ,', ',')
    rev_strng = rev_strng.replace(' .', '.')
    rev_strng = rev_strng.replace("\' ", "'")
    rev_strng = rev_strng.replace(" \'", "'")
    review_lst1.append(rev_strng)
```

```python
len(review_lst1)
```

```python
for review_pos in pos_rev:
    review_text_pos = nltk.corpus.movie_reviews.words(review_pos)
    rev_strng = " ".join(review_text_pos)
    rev_strng = rev_strng.replace(' ,', ',')
    rev_strng = rev_strng.replace(' .', '.')
    rev_strng = rev_strng.replace("\' ", "'")
    rev_strng = rev_strng.replace(" \'", "'")
    review_lst1.append(rev_strng)
```

```python
len(review_lst1)
```

```python
# let's create targets
pos_target = np.ones((1000), dtype=np.int)
neg_target = np.zeros((1000), dtype=np.int)
```

```python
# now put both target in a list
target_lst = []
for neg_tar in neg_target:
    target_lst.append(neg_tar)

for pos_tar in pos_target:
    target_lst.append(pos_tar)
```

```python
len(target_lst)
```

```python
# Now change this list into series
y = pd.Series(target_lst)
y.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vct = CountVectorizer(lowercase=True, stop_words='english', min_df=2)
```

```python
X_cvt = count_vct.fit_transform(review_lst1)
```

```python
X_cvt.shape
```

```python
X_names = count_vct.get_feature_names()
X_names
```

```python
#Now create a dataframe
X_cvt = pd.DataFrame(X_cvt.toarray(), columns=X_names)
X_cvt.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_cvt, y, test_size=0.30, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
pred = gnb.fit(X_train,y_train).predict(X_test)
```

###### Let's try multinomialNB too

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
clf = MultinomialNB()
```

```python
clf.fit(X_train,y_train)
```

```python
prediction = clf.predict(X_test)
type(prediction)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

#### Classification Report and Confusion Matrix for GuassianNB

```python
print(classification_report(y_test, pred))
print('\n')
print(confusion_matrix(y_test, pred))
```

#### Classification Report and Confusion Matrix for MultinomialNB

```python
print(classification_report(y_test, prediction))
print('\n')
print(confusion_matrix(y_test, prediction))
```

We can see the The accuracy of MultinomialNB is far better than GuassianNB.
